#Importing Libraries

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

#Loading Model

```python
cnn = tf.keras.models.load_model('/content/drive/MyDrive/models/fruits_veg_trained_model.h5')
```

##Single prediction

```python
import cv2
image_path='/content/drive/MyDrive/input/test/onion/Image_10.jpg'
img = cv2.imread(image_path)
plt.imshow(img)
plt.title('Test image')
```

#Testing Model

```python
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr]) # Converting single image to batch
prediction = cnn.predict(input_arr)
```

```python
print(prediction)
```

```python
test_set=tf.keras.utils.image_dataset_from_directory('/content/drive/MyDrive/input/test',
                                                         labels='inferred',
                                                         label_mode='categorical',
                                                         class_names=None,
                                                         color_mode='rgb',
                                                         batch_size=32,
                                                         image_size=(64,64),
                                                         shuffle =True,
                                                         seed=None,
                                                         validation_split=None,
                                                         subset=None,
                                                         interpolation='bilinear',
                                                         follow_links=False,
                                                         crop_to_aspect_ratio=False
                                                         )
```

```python
test_set.class_names
```

```python
result_index = np.where(prediction[0]==max(prediction[0]))
print(result_index[0][0])
```
